```python
import numpy as np
import pandas as pd
fn1 = "/mnt/SSD5/yipeng/VAE/res/2023-12-28_1/fold_0/train_81.npz"
fn2 = "/mnt/SSD5/yipeng/VAE/res/2023-12-28_2/fold_0/train_81.npz"
def load_data(fn):
    res = {}
    data = np.load(fn)
    for key in data.files:
        value = data[key]
        if len(value.shape) == 1:
            res[key] = value
            continue
        for i in range(len(value[0])):
            res[key + "_" + str(i)] = value[:,i]
    df = pd.DataFrame(res)
    return df
df_f1 = load_data(fn1)
df_f2 = load_data(fn2)
```
